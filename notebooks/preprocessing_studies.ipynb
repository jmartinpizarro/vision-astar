```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
PREPROCESSED_DATA_ROUTE = "../data/cellnet_preprocessed_data/preprocessed_dataset.csv"

df = pd.read_csv(PREPROCESSED_DATA_ROUTE)
df.info()
```

```python
def render_classes_distribution(df):
    # Todas las clases
    counts = df["label"].value_counts()
    plt.bar(counts.index, counts.values)
    plt.xlabel("Classes")
    plt.ylabel("Frequency")
    plt.title("Distribution of all classes")
    plt.show()

    # now render the ones which are not " "
    df_filtered = df[df["label"] != " "]
    counts_filtered = df_filtered["label"].value_counts()
    plt.bar(counts_filtered.index, counts_filtered.values)
    plt.xlabel("Filtered Classes")
    plt.ylabel("Frequency")
    plt.title("Distribution without ' '")
    plt.show()
```

```python
render_classes_distribution(df)
```

Clearly it is possible to see that the dataset is not balanced. For improving this, and avoiding problems such as overfitting, low learning or same validation accuracy, it is necessary to increase the size of the dataset (for the X, O and G classes).
